```python
# calculate the percentage of your 10th mark
def percentage():
    s1=int(input("Enter s1  "))
    s2=int(input("Enter  s2"))
    s3=int(input("Enter : s3"))
    s4=int(input("Enter : s4"))
    s5=int(input("Enter  s5"))
    tot=s1+s2+s3+s4+s5
    perc=tot/5
    return perc
    
```
